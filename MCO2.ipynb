## $n^{2}$ Algorithms

```python
# to be used for the three n^2 algorithm
def swap(elem1, elem2):
  return elem2, elem1
```

### Bubble Sort

```python
def bubbleSort(numbers):
  freq_count = 0
  arr = copy.deepcopy(numbers)
  for i in range (len(numbers)-1, 0, -1):
    for j in range(1, i+1):
      freq_count+=1
      if (arr[j-1] > arr[j]):
        arr[j-1], arr[j] = swap(arr[j-1], arr[j])
  return freq_count, len(numbers)
```

### Insertion Sort

```python
def insertionSort(numbers):
  freq_count = 0
  arr = copy.deepcopy(numbers)
  for i in range (1, len(numbers)):
    key = arr[i]
    j=i-1
    while j >= 0 and arr[j] > key:
      freq_count+=1
      arr[j+1] = arr[j]
      j-=1
    freq_count+=1
    arr[j+1] = key
  return freq_count, len(numbers)
```

### Selection Sort

```python
def selectionSort(numbers):
  freq_count = 0
  arr = copy.deepcopy(numbers)
  for i in range(len(numbers)):
    m = i
    for j in range(i+1, len(numbers)):
      freq_count+=1
      if (arr[j] < arr[m]):
        m = j;
    arr[i], arr[m] = swap(arr[i], arr[m])
  return freq_count, len(numbers)
```

### Merge Sort

```python
# Merge sort algorithm
def merge(left, right, freq_count):
  # store new list
  new_list = []

  # indices of left and right half
  left_p = 0
  right_p = 0

  while left_p < len(left) or right_p < len(right):
    freq_count+=1
    # if left half is out of elements
    if left_p == len(left):
      # append right element
      new_list.append(right[right_p])
      right_p += 1 # increment right pointer
    # if right half is out of elements
    elif right_p == len(right):
      # append left element
      new_list.append(left[left_p])
      left_p += 1 # increment left pointer
    # if element in left half is less than element in right half
    elif left[left_p] < right[right_p]:
      new_list.append(left[left_p])
      left_p += 1 # increment left pointer
    # if element in right half is less than element in left half
    else:
      new_list.append(right[right_p])
      right_p += 1 # increment right pointer
  freq_count+=1
  return new_list, freq_count

def merge_sort(numbers, freq_count):
  # if there is more than one element
  if len(numbers) > 1:
    mid = len(numbers) // 2
  
    # if even, shift mid to the left
    if len(numbers) % 2 == 0:
      mid -= 1

    left = numbers[:mid + 1]
    right = numbers[mid + 1:]

    # sort both halves
    left, freq_count = merge_sort(left, freq_count)
    right, freq_count = merge_sort(right, freq_count)

    # merge both halves
    numbers, freq_count = merge(left, right, freq_count)

  return numbers, freq_count
```

### Quick Sort

```python
# Quick sort algorithm
def partition(numbers, freq_count):
    # initialization
    pivot_ind = 0
    trash_collector = 1
    dump_index = pivot_ind
    pivot = numbers[pivot_ind]

    # for each element to the right of the pivot
    for trash_collector in range(1, len(numbers)):
        freq_count+=1
        if numbers[trash_collector] < pivot:
            # move dump index ahead
            dump_index += 1

            # swap trash and dump index
            numbers[trash_collector], numbers[dump_index] = numbers[dump_index], numbers[trash_collector]

    # place pivot in the correct spot
    numbers[pivot_ind], numbers[dump_index] = numbers[dump_index], numbers[pivot_ind]

    return numbers, dump_index, freq_count

def quicksort(numbers, freq_count):
    # if at least 2 elements
    if len(numbers) > 1:
        # get the partitioned array numbers AND the pivot index p
        numbers, p, freq_count = partition(numbers, freq_count)

        # if pivot is at the start
        if p == 0:
            left = []
        else:
            left = numbers[:p]

        # if pivot is at the end
        if p + 1 == len(numbers):
            right = []
        else:
            right = numbers[p + 1:]

        left, freq_count = quicksort(left, freq_count)
        right, freq_count = quicksort(right, freq_count)

        # concatenate left half, list w/ pivot, and right
        numbers = left + [numbers[p]] + right
    return numbers, freq_count
```

### Shell Sort

```python
# Reference: https://www.programiz.com/dsa/shell-sort
def shellSort(arr):
  # Rearrange elements at each n/2, n/4, n/8, ... intervals
  freq_count = 0
  array = copy.deepcopy(arr)
  interval = len(array) // 2
  while interval > 0:
    for i in range(interval, len(array)):
      temp = array[i]
      j = i
      while j >= interval and array[j - interval] > temp:
        freq_count+=1
        array[j] = array[j - interval]
        j -= interval
      freq_count+=1
      array[j] = temp
    interval //= 2
  return freq_count, len(array)
```
